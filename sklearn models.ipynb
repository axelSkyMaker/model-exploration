```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=500, centers=4,
                       cluster_std=0.8, random_state=0)


#  Code from http://www.cse.chalmers.se/~richajo/dit866/lectures/l3/Plotting%20decision%20boundaries.html
def plot_decision_boundary(clf, X, Y, cmap='Paired_r'):
    h = 0.02
    x_min, x_max = X[:,0].min() - 10*h, X[:,0].max() + 10*h
    y_min, y_max = X[:,1].min() - 10*h, X[:,1].max() + 10*h
    print(x_min, x_max)
    print(y_min, y_max)
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    print(Z.shape)
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10,10))
    plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.25)
    plt.contour(xx, yy, Z, colors='k', linewidths=0.7)
    plt.scatter(X[:,0], X[:,1], c=Y, cmap=cmap, edgecolors='k');
```

```python
from sklearn.linear_model import Perceptron

clf = Perceptron()
clf.fit(X, y_true)

plot_decision_boundary(clf, X, y_true)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X, y_true)

plot_decision_boundary(clf, X, y_true)
plt.show()
```

```python
points = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
print(points.shape)
tmp_x = points[:, 0]
tmp_y = points[:, 1]

plt.scatter(tmp_x, tmp_y)
X_linear_example = points
y_linear_example = [0, 1, 1, 0]
clf = Perceptron()
clf.fit(X_linear_example, y_linear_example)
pred = clf.predict(X_linear_example)

print(pred)
plot_decision_boundary(clf, X_linear_example, y_linear_example)
```

```python
from sklearn.neural_network  import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
points = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
print(points.shape)
tmp_x = points[:, 0]
tmp_y = points[:, 1]

plt.scatter(tmp_x, tmp_y)
X_linear_example = points
y_linear_example = [0, 1, 1, 0]
clf = KNeighborsClassifier(1)
clf.fit(X_linear_example, y_linear_example)
pred = clf.predict(X_linear_example)

print(pred)
plot_decision_boundary(clf, X_linear_example, y_linear_example)
```
